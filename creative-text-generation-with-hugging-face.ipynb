```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
tokenizer =AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
input_text="Bangladesh is a small country.it has large population"
```

```python
input_ids = tokenizer(input_text, return_tensors="pt")
print(input_ids)
```

```python
input_ids = input_ids['input_ids'].to(device)
```

```python
max_len=160
```

```python
model.to(device)
output = model.generate(input_ids, max_length=max_len, num_beams=5, do_sample=False)
```

```python
output
```

```python
print(tokenizer.decode(output[0]))
```

```python

output = model.generate(input_ids, max_length=max_len, num_beams=5,
                        do_sample=False, no_repeat_ngram_size=2)
tokenizer.decode(output[0])
```

```python

input_text = "say something about Shakib Al Hasan  bangladeshi cricketer"
max_length = 170

input_ids = tokenizer(input_text, return_tensors="pt")
input_ids = input_ids['input_ids'].to(device)

output = model.generate(input_ids, max_length=max_length,
                        do_sample=True, top_p=1)
tokenizer.decode(output[0])
```
